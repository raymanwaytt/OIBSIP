## Extraction from Serving Size column

The Serving Size column consists of different units, depending on the Food Category
(oz, g, fl oz).
This section focus on extracting the numerical values from the Serving size by using a function based on REGEX.

#### Creating a REGEX function to extract values from the Serving Size

```python
# Function to extract the numeric value for different units
def extract_value_by_unit(serving_size, unit):
    # Create the regex pattern for each unit
    pattern = rf'(\d+(\.\d+)?)\s*({unit})'
               
    # Search for the pattern in the serving size
    match = re.search(pattern, serving_size)
    if match:
        return float(match.group(1))  # Return the numeric value as a float
    return None  # If no match, return None
```

#### Creating new Columns for different Serving Size Units and mapping the values using the above function. extract_value_by_unit()

```python
# Create a new column for 'Serving Size (g)' and extract gram values
df['Serving Size (g)'] = df['Serving Size'].apply(lambda x: extract_value_by_unit(x, 'g'))

# Explanation:
# - The `extract_value_by_unit` function extracts the numeric value for the specified unit ('g' in this case).
# - The result is stored in a new column called 'Serving Size (g)'.

# Extract fluid ounces ('fl oz') values, convert to milliliters (ml), and store in a new column
df['Serving Size (ml)'] = df['Serving Size'].apply(lambda x: extract_value_by_unit(x, 'fl oz')) * 29.57

# Explanation:
# - The `extract_value_by_unit` function extracts the numeric value for 'fl oz'.
# - Since 1 fluid ounce (fl oz) = 29.57 milliliters (ml), we multiply the extracted value by 29.57 to convert it to ml.
# - The result is stored in a new column called 'Serving Size (ml)'.
```

## Calorie Density

This section delves into creating calorie density:
- calories per gram (g) for each solid food item 
- calories per centilitre (cL) for each liquid item

```python
# creating the Calories per gram column for the solids df
df['Calorie_density (g)'] = round(df['Calories']/df['Serving Size (g)'], 2)
```

```python
# creating the Calories per gram column for the solids df
df['Calories_density (ml) '] = round(df['Calories']/df['Serving Size (ml)'], 2)
```

##  Crispy items vs. Grilled items : calories and fat

```python
# Filter items containing "crispy" or "grilled"
crispy_items = df[df['Item'].str.contains('Crispy', case=False, na=False)]
grilled_items = df[df['Item'].str.contains('Grilled', case=False, na=False)]

# Calculate average calories and fat for crispy and grilled items
crispy_avg_calories = crispy_items['Calories'].mean()
crispy_avg_fat = crispy_items['Total Fat'].mean()

grilled_avg_calories = grilled_items['Calories'].mean()
grilled_avg_fat = grilled_items['Total Fat'].mean()

# Data for visualization
labels = ['Calories', 'Total Fat']
crispy_stats = [crispy_avg_calories, crispy_avg_fat]
grilled_stats = [grilled_avg_calories, grilled_avg_fat]

x = range(len(labels))
```

```python
# Set positions for bars on the x-axis
x = range(len(labels))  # Number of labels determines x positions

# Create figure and axis with a larger size for clarity
fig, ax = plt.subplots(figsize=(12, 7))

# Create bars for Crispy and Grilled categories
bars1 = ax.bar(x, crispy_stats, width=0.4, label='Crispy', color='skyblue', align='center')
bars2 = ax.bar([i + 0.4 for i in x], grilled_stats, width=0.4, label='Grilled', color='salmon', align='center')

# Add value labels on top of each bar
for bar in bars1 + bars2:
    height = bar.get_height()  # Get bar height
    ax.annotate(f'{height:.1f}',  # Convert to 1 decimal place
                xy=(bar.get_x() + bar.get_width() / 2, height),  # Position on top of the bar
                xytext=(0, 6),  # Slightly offset text above the bar
                textcoords="offset points",
                ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')

# Improve x-axis readability
ax.set_xticks([i + 0.2 for i in x])  # Shift x-ticks to center labels correctly
ax.set_xticklabels(labels, fontsize=11)  
# Add labels and title
ax.set_ylabel('Average Values', fontsize=12)
ax.set_title('Crispy vs Grilled: Calories & Fat Comparison', fontsize=14) 
# Add legend with a better position
ax.legend(loc='upper right', fontsize=11)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

## Beverages Caloric compared to other Food Categories

```python
# Define a list of food categories excluding beverages
others = ['Breakfast', 'Beef & Pork', 'Chicken & Fish', 'Salads', 'Snacks & Sides', 'Desserts']

# Separate the data into beverages and other food categories
beverages = df[df['Category'] == 'Beverages']  # Filter rows where the category is 'Beverages'
other_food = df[df['Category'].isin(others)]  # Filter rows for all other specified categories

# Calculate total calories for beverages and other food items
beverage_total_calories = beverages['Calories'].sum()  # Total calories from beverages
other_food_total_calories = other_food['Calories'].sum()  # Total calories from other food items

# Calculate the total calorie count across the entire dataset
total_calories = df['Calories'].sum()

# Compute the percentage contribution of each category to total calories
beverage_percentage = (beverage_total_calories / total_calories) * 100  # Percentage from beverages
other_food_percentage = (other_food_total_calories / total_calories) * 100  # Percentage from other food
```

## HEALTHINESS SCORE
This section focuses on creating a Healthiness Score Column, calculating a 'Healthiness Score' based on calorie, sodium, and sugar content

### Each factor is weighted: 
- Calories (40% weight) relative to a 2000-calorie daily intake
- Sodium (40% weight) relative to the 2300mg recommended daily limit
- Sugar (20% weight) relative to a 50g daily limit
#### (Weights align with USDA dietary priorities)

### Effects
-  Excess calories can lead to weight gain and other health issues.
- Sodium (40% weight) :  High sodium levels are linked to high blood pressure and heart disease.
- Sugar (20% weight) :  Too much sugar increases the risk of diabetes and obesity.

#### Lower scores indicate healthier items, while higher scores indicate less healthy ones.

```python
# Calculate a 'Healthiness Score' based on calorie, sodium, and sugar content

df['Healthiness_Score'] = round(
    (
        (df['Calories'] / 2000 * 0.4) +  # Calories contribution (40% weight)
        (df['Sodium'] / 2300 * 0.4) +  # Sodium contribution (40% weight)
        (df['Sugars'] / 50 * 0.2)  # Sugar contribution (20% weight)
    ), 
    2  # Round to 2 decimal places for readability
)
```

```python
metrics = ['Item', 'Category', 'Serving Size', 'Calories', 'Total Fat', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Healthiness_Score']
```

### Most Healthy Items

```python
# checks for items where Healthiness_Score is 0
df[df['Healthiness_Score'] == 0][metrics]
```

#### Healthiest Item:

Water, Diet Coke & Items in Coffee & Tea Category, with a healthiness score of 0.0. It contains no calories, fat, sugar, or sodium, making it an ideal choice for health-conscious customers.

### Top 10 most Unhealthy Items

```python
# Sort the DataFrame by Healthiness_Score in descending order
top_unhealthy = df.sort_values(by='Healthiness_Score', ascending=False).head(10)

# Display the top 10 most unhealthy items
top_unhealthy[metrics]
```

#### Least Healthy Item:

Chicken McNuggets (40-piece) is the least healthy, with an extraordinarily high healthiness score (1.01) 101% of daily limit. Its combination of high calories (1880), fat (118g), and sodium (3600mg) positions it as the most indulgent and unhealthy option on the menu.
