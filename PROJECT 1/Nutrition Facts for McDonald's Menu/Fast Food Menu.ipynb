## Extraction from Serving Size column

The Serving Size column consists of different units, depending on the Food Category
(oz, g, fl oz).
This section focus on extracting the numerical values from the Serving size by using a function based on REGEX.

```python
# Function to extract the numeric value for different units
def extract_value_by_unit(serving_size, unit):
    # Create the regex pattern for each unit
    pattern = rf'(\d+(\.\d+)?)\s*({unit})'
    
    # Search for the pattern in the serving size
    match = re.search(pattern, serving_size)
    if match:
        return float(match.group(1))  # Return the numeric value as a float
    return None  # If no match, return None
```

#### Creating new Columns for different Serving Size Units and mapping the values using the above function.

```python
# Create new column for gram(g) and input the right Serving Size values
df['Serving Size (g)'] = df['Serving Size'].apply(lambda x: extract_value_by_unit(x, 'g')) # extract g values

# Extract (ml & fl oz) values, convert to cL, then map to cL column
df['Serving Size (cL)'] = df['Serving Size'].apply(lambda x: extract_value_by_unit(x, 'ml')) * 0.1 # Extract ml, converts to cL
df['Serving Size (cL)'] = df['Serving Size'].apply(lambda x: extract_value_by_unit(x, 'fl oz')) * 2.95735 # Extract 'fl oz', convert to cL
```

```python
# round off the 'Serving Size (cL)' to 2.d.p
df['Serving Size (cL)'] = round(df['Serving Size (cL)'], 2)
```

## Calorie Density

This section delves into creating calorie density:
- calories per gram (g) for each solid food item 
- calories per centilitre (cL) for each liquid item

```python
# creating the Calories per gram column for the solids df
df['Calories per gram'] = round(df['Calories']/df['Serving Size (g)'], 2)
```

```python
# creating the Calories per gram column for the solids df
df['Calories per Centilitre '] = round(df['Calories']/df['Serving Size (cL)'], 2)
```

##  Crispy items vs. Grilled items : calories and fat

```python
# Filter items containing "crispy" or "grilled"
crispy_items = df[df['Item'].str.contains('Crispy', case=False, na=False)]
grilled_items = df[df['Item'].str.contains('Grilled', case=False, na=False)]

# Calculate average calories and fat for crispy and grilled items
crispy_avg_calories = crispy_items['Calories'].mean()
crispy_avg_fat = crispy_items['Total Fat'].mean()

grilled_avg_calories = grilled_items['Calories'].mean()
grilled_avg_fat = grilled_items['Total Fat'].mean()

# Data for visualization
labels = ['Calories', 'Total Fat']
crispy_stats = [crispy_avg_calories, crispy_avg_fat]
grilled_stats = [grilled_avg_calories, grilled_avg_fat]

x = range(len(labels))
```

```python
# Plot
plt.bar(x, crispy_stats, width=0.4, label='Crispy', align='center')
plt.bar([i + 0.4 for i in x], grilled_stats, width=0.4, label='Grilled', align='center')

plt.xticks([i + 0.2 for i in x], labels)
plt.ylabel('Average Values')
plt.title('Crispy vs Grilled: Calories & Fat Comparison')
plt.legend()
plt.show()
```

## Beverages Caloric compared to other Food Categories

```python
others = ['Breakfast', 'Beef & Pork', 'Chicken & Fish', 'Salads', 'Snacks & Sides', 'Desserts']

beverages = df[df['Category'] == 'Beverages']
other_food = df[df['Category'].isin(others)]

# Total calories
beverage_total_calories = beverages['Calories'].sum()
other_food_total_calories = other_food['Calories'].sum()

# Total calories in the dataset
total_calories = df['Calories'].sum()

# Percentage contribution
beverage_percentage = (beverage_total_calories / total_calories) * 100
other_food_percentage = (other_food_total_calories / total_calories) * 100
```

```python
# Data for visualization
categories = ['Beverages Calories', 'Other Foods Calories']
contributions = [beverage_percentage, other_food_percentage]

# Pie chart
plt.figure(figsize=(10,8))
plt.pie(contributions, labels=categories, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'])
plt.title('Caloric Contribution: Beverages vs Other Food')
plt.show()
```
